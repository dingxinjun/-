```python
import numpy as np
import matplotlib.pyplot as plt

budyko_curve_x = np.arange(0, 3, 0.05)
energy_limit_x = np.arange(0, 1.0001, 0.05)
x = np.arange(0, 1.0001, 0.05)

budyko_curve_y = np.power((budyko_curve_x*np.tanh(1/budyko_curve_x)*(1-np.exp(-budyko_curve_x))),0.5)

water_limit_y = 1+budyko_curve_x*0
energy_limit_y = energy_limit_x
y = 1 + x*0

plt.plot(budyko_curve_x,budyko_curve_y,  linestyle='--')

plt.plot(energy_limit_y,energy_limit_x, c='r')
plt.plot(budyko_curve_x,water_limit_y,c='b')
plt.plot(y,x,linestyle='-.', c='k',label='_nolegend_')

plt.ylabel("Actual ET/P")
plt.xlabel("Potential ET/P")
plt.title('Budyko Curve')
plt.grid(b=True, which='major', color='#666666', linestyle='-',alpha=0.2)
# Show the minor grid lines with very faint and almost transparent grey lines
plt.minorticks_on()
plt.legend(["Budyko Curve","Energy Limit", "Water Limit"])
# plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
budyko_curve_x = np.arange(0, 3, 0.05)
budyko_curve_y = np.power((budyko_curve_x*np.tanh(1/budyko_curve_x)*(1-np.exp(-budyko_curve_x))),0.5)
energy_limit_x = np.arange(0, 1.0001, 0.05)
water_limit_x=np.arange(1,3,0.05)

water_limit_y = 1+water_limit_x*0
energy_limit_y = energy_limit_x

x = np.arange(0, 1.0001, 0.05)
y = 1 + x*0

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim(0, 3)#600
ax.set_ylim(0, 1.2)
#绘制bukyko曲线x区间为[0,3]
ax.plot(budyko_curve_x,budyko_curve_y,  linestyle='--')
#绘制能量限制阶段，区间为（0，1）
ax.plot(energy_limit_x,energy_limit_y, c='r')
#绘制水量限制阶段，区间为(1,3)
ax.plot(water_limit_x,water_limit_y,c='b')
#这里ax.text(x坐标，y坐标，string，)
ax.text(2,0.8, 'Budyko curve',ha="center", va="center", fontsize = 20,color='k')
ax.text(2,1.03, 'water limit',ha="center", va="center", fontsize = 20,color='blue')
ax.text(0.7,0.8, 'energy limit',ha="center", va="center",rotation=60, fontsize = 20,color='red')
# 黑线
ax.plot(y,x,linestyle='-.', c='k',label='_nolegend_')
def budyko(DI,W):
    F_fu=1+DI-((1+DI**W)**(1/W))
    return F_fu
for w in range(1,30,3):
    budyko_curve_y =budyko(budyko_curve_x,w)
    ax.plot(budyko_curve_x,budyko_curve_y,  linestyle='-',label='_nolegend_')
    
ax.legend(["Budyko Curve","Energy Limit", "Water Limit"])

ax.set_ylabel("Actual ET/P")
ax.set_xlabel("Potential ET/P")
ax.set_title('Budyko Curve')
#添加底色
ax.axvspan(0,1.54,facecolor='dodgerblue',edgecolor='k',alpha=0.2)
ax.axvspan(1.54,2,facecolor='lightskyblue',edgecolor='r',alpha=0.5)
ax.axvspan(2,3,facecolor='skyblue',alpha=0.2)
# ax.axvspan(5,10,facecolor='lightblue')

plt.grid(False)
plt.show()
```
