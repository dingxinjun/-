```python
import numpy as np
import statsmodels.api as sm # recommended import according to the docs
import matplotlib.pyplot as plt

NSE_PDM = np.random.uniform(0, 1, 50)
res_freq1 = stats.relfreq(NSE_PDM, numbins=1000) # numbins 是统计一次的间隔(步长)是多大
pdf_value1 = res_freq1.frequency #概率密度PDF
cdf_value1 = np.cumsum(res_freq1.frequency) #累积分布CDF
x1 = res_freq1.lowerlimit + np.linspace(0, res_freq1.binsize * res_freq1.frequency.size, res_freq1.frequency.size)
exceed_P1 = 1 - cdf_value1
plt.plot(exceed_P1,x1,linewidth=1,label = 'PDM')
plt.show()
```
